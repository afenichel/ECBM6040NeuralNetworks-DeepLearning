#### Creating plots for errors on the CIFAR10 tests with 30 epochs

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy
%matplotlib inline
```

```python
perf_files = ['LRout_cifar10_errors.csv', 
              'svmout_cifar10_errors.csv',
              'LRout_bin_stochastic_cifar10_errors.csv', 
              'LRout_bin_deterministic_cifar10_errors.csv',
              'LRout_bin_stochastic_cifar10_errors.csv',
              'svmout_bin_deterministic_cifar10_errors.csv']

perf_filenames = ['Logistic Output Layer without Binary Connect',
                  'SVM Output Layer without Binary Connect',
                  'Logistic Output Layer with Stochastic Binary Connect',
                  'Logistic Output Layer with Deterministic Binary Connect',
                  'SVM Output Layer with Stochastic Binary Connect',
                  'SVM Output Layer with Deterministic Binary Connect'
                 ]
pdic = {}
for i, pfile in enumerate(perf_files):
    pdic['m'+str(i)] = pd.read_csv('output/'+pfile)
```

```python
for i, pfile in enumerate(perf_filenames):
    print(pfile)
```

```python
n = 1000
for i, pfile in enumerate(perf_filenames):
    m = 'm'+str(i)
    plt.figure(figsize=(8,5))
    plt.plot(pdic[m].epoch[0:n], pdic[m].test_score[0:n], label='Test')
    plt.plot(pdic[m].epoch[0:n], pdic[m].this_validation_loss[0:n], label='Validation')
    plt.plot(pdic[m].epoch[0:n], pdic[m].this_training_loss[0:n], label='Training')
    plt.title("CIFAR10 Model performance \n"+pfile)
    plt.ylabel('Classification Error (%)')
    plt.xlabel('Number of epochs')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
for (filex, filename) in zip(perf_filenames, perf_files):
    df = pd.read_csv('output/'+filename)
    print filex,'had a test error of' , df.ix[df.this_validation_loss.argmin(),'test_score']*100
```
