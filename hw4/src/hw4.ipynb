#### Bullet 4: Experiment with word embeddings

Here I look at differences in word embeddings that are similar, as well as embeddings of specific words that are similar.

##### Similarities between differences of word embeddings

```python
emb, idx2label, idx2word= test_rnnslu(nhidden=200, win=5, emb_dimension=100, verbose=True, nepochs=25, normal=True)
```

```python
from scipy import spatial

def diff_matrix(r1,r2,r3,r4):
    diff={}
    for i in range(r1,r2):
        for j in range(r3,r4):
            if i<j:
                diff[(i,j)]=numpy.abs(numpy.asarray(emb[i])-numpy.asarray(emb[j]))
        
    diff_mat=numpy.zeros((len(diff), len(diff)))+numpy.inf
    for n, i in enumerate(diff.keys()):
        for m, j in enumerate(diff.keys()[n+1:]):
            diff_mat[n, n+m+1]=spatial.distance.euclidean(diff[i], diff[j])
        
    for n, i in enumerate(numpy.argmin(diff_mat,0)):
        if diff_mat[i,n]<0.75:
            if ('DIGIT' not in idx2word[diff.keys()[n][0]]) and ('DIGIT' not in idx2word[diff.keys()[i][0]]) and ('DIGIT' not in idx2word[diff.keys()[i][1]]) and ('UNK' not in idx2word[diff.keys()[i][0]]) and ('UNK' not in idx2word[diff.keys()[i][1]]):
                if (diff.keys()[n][0]!=diff.keys()[i][0]) and (diff.keys()[n][1]!=diff.keys()[i][1]) and (diff.keys()[n][1]!=diff.keys()[i][0]) and (diff.keys()[n][0]!=diff.keys()[i][1]):
                    print '\"%s\" minus \"%s\" approx. equals \"%s\" minus \"%s\" with distance %s' % (idx2word[diff.keys()[n][0]], idx2word[diff.keys()[n][1]], idx2word[diff.keys()[i][0]], idx2word[diff.keys()[i][1]], diff_mat[i,n]) 
```

```python
diff_matrix(10,50,30,70)
```

```python
diff_matrix(200,250,300,350)
```

```python
diff_matrix(100,150,200,250)
```

```python
diff_matrix(75,125,225,275)
```

##### Similarities between word embeddings

```python
from scipy import spatial
nwords=emb.shape[0]
dist=numpy.zeros((nwords, nwords))
for i in range(nwords):
    for j in range(nwords):
        if i!=j:
            dist[i,j]=spatial.distance.euclidean(emb[i], emb[j])
        elif i==j:
            dist[i,j]=numpy.inf
```

```python
print 'Words that are withing 0.5 euclidean distance of each other:'
for n, i in enumerate(numpy.argmin(dist,0)):
    if dist[n,i]<0.5:
        print idx2word[n],'is close to', idx2word[i],'with distance', dist[n,i]
```
