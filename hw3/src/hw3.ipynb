#### Bullet 3: Explain why the back-propogation algorithm works on ms-cnn

Back propogation takes the sum of all the derivatives of a node with respect to all other nodes that feed into it. Because the chain rule is required to back propogate through the whole graph, if our layer0 feeds into layer1, and a combination of layer0 and layer1 feed into layer2, we can still take the gradient with respect to each layer's parameters. 

The gradient of layer2 with respect to layer0 paramters is then a sum of the derivative of the input of layer2 w.r.t. layer0 params and the derivative using the chain rule of the input of layer2 with respect to layer1 params, times the derivative of layer1 with respect to layer0 params. We also take the derivative of layer2 input with respect to layer1 parameters. Wherever a gradient is taken on a child node of layer 2 this calculation will be performed within the chain rule.
